# Triplet Network

### Imports

```python
from IPython.display import Image, SVG
%matplotlib inline

import seaborn as sns

import time
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from sklearn.utils import shuffle
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, Normalizer
from sklearn.metrics import roc_curve, auc, accuracy_score, log_loss
from sklearn.model_selection import cross_val_predict
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

import keras
import keras.backend as K
from keras.models import Model, Sequential
from keras.layers import Input, Dense, LeakyReLU, Conv2D, RepeatVector, Flatten, Reshape, Lambda, Conv2DTranspose, BatchNormalization, concatenate
from keras.regularizers import l2, l1
from keras.constraints import UnitNorm, MaxNorm

config = tf.ConfigProto()
config.gpu_options.allow_growth=True
sess = tf.Session(config=config)
```

## Hyperparameter & Oprions

```python
### Preprocessing
bkg_filename = "../background_Full_Test3.txt"
sig_filename = "../signal_Full_Test3.txt"
drop_PHI_columns = False
n_train = 24000
n_test = 3000
scaler = MinMaxScaler(feature_range=(0,1))
#scaler = Normalizer()

### Network -- General
enc_dim = 3
L_distance = 0.5
useFlat = False
out_activation = 'sigmoid'

### Network -- Flat
n_layers = 31
h_activation = 'selu'  # 'leaky' for LeakyReLU
kconstraint = None #UnitNorm()
bconstraint = None #UnitNorm()
areg = None #l2(1e-6)
kreg = None #l1(0.001)

### Network -- Convolutional
n_conv = 6
leak = 0.2
min_filters = 32

### Training
n_epochs = 50
batch_size = 512
adam = keras.optimizers.Adam(
    lr=1e-5, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
```

### Data Loading

```python
bkg_all = pd.read_csv(bkg_filename, delimiter=' ', index_col=False)
sig_all = pd.read_csv(sig_filename, delimiter=' ', index_col=False)

def coordinate_change (df):
    pt1 = np.sqrt(df['px1']**2 + df['py1']**2)
    theta1 = np.arctan2(pt1,df['pz1'])
    eta1 = -1 * np.log(np.tan(theta1/2))
    phi1 = np.arctan2(df['py1'],df['px1'])
    pt2 = np.sqrt(df['px2']**2 + df['py2']**2)
    theta2 = np.arctan2(pt2,df['pz2'])
    eta2 = -1 * np.log(np.tan(theta2/2))
    phi2 = np.arctan2(df['py2'],df['px2'])
    df['px1'] = pt1
    df['py1'] = eta1
    df['pz1'] = phi1
    df['px2'] = pt2
    df['py2'] = eta2
    df['pz2'] = phi2
    df.rename(columns={'px1':'$p_{t1}$', 'py1':'$\eta_{1}$', 'pz1':'$\phi_{1}$', 'px2':'$p_{t2}$',
                       'py2':'$\eta_{2}$', 'pz2':'$\phi_{2}$', 'E1':'$E_1$', 'E2':'$E_2$', 'M1':'$M_1$',
                       'M2':'$M_2$', 'M12':'$M_{1 2}$'}, inplace=True)
    
coordinate_change(sig_all)
coordinate_change(bkg_all)

if drop_PHI_columns:
    sig_sel = sig_all.drop(columns = ['$\phi_{1}$', '$\phi_{2}$'])
    bkg_sel = bkg_all.drop(columns = ['$\phi_{1}$', '$\phi_{2}$'])
else:
    sig_sel = sig_all
    bkg_sel = bkg_all
    
data_header = list(sig_sel)

sig_shuffled = scaler.fit_transform(shuffle(sig_sel))
bkg_shuffled = scaler.transform(shuffle(bkg_sel))
input_dim = sig_shuffled.shape[1]
```

### Preprocessing

```python
[sig_train, bkg_train] = [np.array(bkg_sel.iloc[:n_train]), np.array(sig_sel.iloc[:n_train])]
[sig_test, bkg_test] = [np.array(bkg_sel.iloc[n_train:2*n_train].reset_index(drop=True)), np.array(sig_sel.iloc[n_train:2*n_train].reset_index(drop=True))]
x_test_tc = np.concatenate([sig_test[:n_test//2], bkg_test[:n_test//2]])
x_test_v = np.concatenate([sig_test[n_test//2:n_test], bkg_test[n_test//2:n_test]])

y_class = np.concatenate([np.repeat(1,n_test//2), np.repeat(0,n_test//2)])


classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="rbf", C=0.025, probability=True),
    NuSVC(probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()]

def generate_triplets(sig, bkg, validsize=0):
    trainsize = 1-validsize

    triplets_train = []
    triplets_valid = []
    
    PositiveS = sig_train[:n_train//3]
    AnchorS = sig_train[n_train//3:2*n_train//3]
    NegativeB = bkg_train[:n_train//3]
    
    PositiveB = bkg_train[n_train//3:2*n_train//3]
    AnchorB = bkg_train[2*n_train//3:n_train]
    NegativeS = sig_train[:2*n_train//3]
    
    Positive = np.concatenate([PositiveS,PositiveB])
    Anchor = np.concatenate([AnchorS, AnchorB])
    Negative = np.concatenate([NegativeB, NegativeS])
    
    n=int((2*n_train//3)*trainsize)
    for i in range(n):
        triplets_train.append([Positive[i], Anchor[i], Negative[i]])
        
    for i in range(2*n_train//3-n):
        triplets_valid.append([Positive[i], Anchor[i], Negative[i]])
                   
    return np.array(triplets_train), np.array(triplets_valid)



x_train, x_valid = generate_triplets(sig_train, bkg_train, validsize=0)

n_input = x_train.shape[0]
n_valid = x_valid.shape[0]
```

### Supervised Classification, pre Metric Change

```python
def try_classification(x_train, x_test, y):

    log_cols=["Classifier", "Accuracy"]
    log = pd.DataFrame(columns=log_cols)

    for clf in classifiers:
        clf.fit(x_train, y)
        name = clf.__class__.__name__
    
        train_predictions = clf.predict(x_test)
        acc = accuracy_score(y, train_predictions)
    
        log_entry = pd.DataFrame([[name, acc*100]], columns=log_cols)
        log = log.append(log_entry)
    return log
```

### Network and Loss 

```python
def create_N_flat(output_dim, n_layers):
    model = Sequential ()
    if (n_layers>1):
        if (h_activation == 'leaky'):
            model.add(Dense(input_dim, input_shape=(input_dim, ), activation='linear', name='N_h1', bias_constraint=bconstraint, kernel_constraint=kconstraint, activity_regularizer=areg, kernel_regularizer=kreg))
            model.add(LeakyReLU())
            for i in range(1, n_layers):
                l_name = "N_h%d" % (i+1)
                model.add(Dense(input_dim, activation='linear', name=l_name, bias_constraint=bconstraint, kernel_constraint=kconstraint, activity_regularizer=areg, kernel_regularizer=kreg))
                model.add(LeakyReLU())
        else:
            model.add(Dense(input_dim, input_shape=(input_dim, ), activation=h_activation, name='N_h1', bias_constraint=bconstraint, kernel_constraint=kconstraint, activity_regularizer=areg, kernel_regularizer=kreg))
            for i in range(1, n_layers-1):
                l_name = "N_h%d" % (i+1)
                model.add(Dense(input_dim, activation=h_activation, name=l_name, bias_constraint=bconstraint, kernel_constraint=kconstraint, activity_regularizer=areg, kernel_regularizer=kreg))
    model.add(Dense(output_dim, activation=out_activation, name="N_h%d"%(n_layers)))
    return model

def create_CNN(output_dim, n_conv):
    model = Sequential()
    model.add(Dense(128, input_shape=(input_dim, ), activation='linear', name='expand'))
    model.add(LeakyReLU(alpha=leak))
    model.add(Reshape((8,8,2), name='reshape'))
    
    for i in range(1, n_conv):
        flt = (n_conv-i+1)*min_filters
        l_name = "conv_%d" % (i)
        model.add(Conv2D(flt, kernel_size=3, strides=1, padding="same", name=l_name))
        model.add(LeakyReLU(alpha=leak))
        
    model.add(Flatten())
    model.add(Dense(128, activation='linear', name='flatten'))
    model.add(LeakyReLU(alpha=leak))
    model.add(Dense(output_dim, activation=out_activation, name='output'))
    
    return model
   
def total_lenght_all(y_true,y_pred):
    return K.maximum(y_pred.shape.as_list()[1],0)
    
def pos_dist(y_true,y_pred):
    total_lenght = y_pred.shape.as_list()[1]
    positive = y_pred[:,0:int(total_lenght*1/3)]
    anchor = y_pred[:,int(total_lenght*1/3):int(total_lenght*2/3)]

    return K.sum(K.square(anchor-positive),axis=1)

def neg_dist(y_true,y_pred):
    total_lenght = y_pred.shape.as_list()[1]
    anchor = y_pred[:,int(total_lenght*1/3):int(total_lenght*2/3)]
    negative = y_pred[:,int(total_lenght*2/3):int(total_lenght*3/3)]

    return K.sum(K.square(anchor-negative),axis=1)

def triplet_loss(y_true,y_pred,d=L_distance):

    total_loss = pos_dist(y_true,y_pred)-neg_dist(y_true,y_pred)+L_distance
 
    return K.maximum(total_loss,0.0)
```

```python
anchor_input = Input((input_dim, ), name='anchor_input')
positive_input = Input((input_dim, ), name='positive_input')
negative_input = Input((input_dim, ), name='negative_input')

if useFlat:
    Shared_N = create_N_flat(enc_dim, n_layers)
else:
    Shared_N = create_CNN(enc_dim, n_conv)

encoded_positive = Shared_N(positive_input)
encoded_anchor = Shared_N(anchor_input)
encoded_negative = Shared_N(negative_input)

TN_out = concatenate([encoded_positive, encoded_anchor, encoded_negative], axis=-1, name='merged_layer')

TN = Model(inputs=[positive_input, anchor_input, negative_input], outputs=TN_out)
TN.compile(loss=triplet_loss, optimizer=adam, metrics=[total_lenght_all, pos_dist, neg_dist])
TN.summary()

Nmetric = Model (anchor_input, encoded_anchor)
```

### Training

```python
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))

history = LossHistory()

print(x_train[:,:,:].shape)
```

```python
y_dummy_train = np.empty((n_input, enc_dim))
y_dummy_valid = np.empty((n_valid, enc_dim))

epochs = range(n_epochs)

TN_train = TN.fit([x_train[:,0,:], x_train[:,1,:], x_train[:,2,:]], y_dummy_train,                    
                epochs=n_epochs,
                batch_size=batch_size,
                shuffle=True,
                #validation_data=([x_valid[:,0,:], x_valid[:,1,:], x_valid[:,2,:]], y_dummy_valid),
                callbacks=[history])
    
```

### Loss & Validation

```python
loss = TN_train.history['loss']
pos_dist = TN_train.history['pos_dist']
neg_dist = TN_train.history['neg_dist']
epochs = range(n_epochs)
lossFig = plt.figure(figsize=(16,8))
plt.rcParams.update({'font.size': 16})
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, pos_dist, 'g', label='Positive distance')
plt.plot(epochs, neg_dist, 'r', label='Negative distance')
plt.plot(epochs, np.repeat(L_distance,n_epochs), linestyle='--',color=(0.6, 0.6, 0.6), label='Margin')
plt.title('Training loss & Metrics')
#plt.yscale('log')
plt.legend()
plt.show()
```
