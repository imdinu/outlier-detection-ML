# AE GAN

### Imports

```python
from IPython.display import Image, SVG
%matplotlib inline

import time
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, Normalizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import cross_val_predict

import keras
from keras.models import Model, Sequential
from keras.layers import Input, Dense, LeakyReLU, Conv2D, RepeatVector, Flatten, Reshape, Lambda, Conv2DTranspose, BatchNormalization, concatenate
from keras import regularizers
from keras.constraints import UnitNorm, MaxNorm

config = tf.ConfigProto()
config.gpu_options.allow_growth=True
sess = tf.Session(config=config)
```

## HYPERPARAMS & OPTIONS

```python
### DATA SELECTION ###
bkg_filename = "../background_Full_Test3.txt"
sig_filename = "../signal_Full_Test3.txt"
drop_PHI_columns = False
n_train = 25000
scaler = MinMaxScaler(feature_range=(0,1))

### OPTIONS ###
savePlots = False
plotsLocation = "./"
skipStep1 = False


### AUTOENCODER ###
encoding_dim = 7
train_steps = 2500
batch_size_d = 512
batch_size_g = 1024
mini_batch = 32
leak=0.2
hidden_activation = 'linear'
encoder_activation = 'sigmoid'
decoder_activation = 'sigmoid'
generator_activation = 'sigmoid'
discriminator_activation = 'sigmoid'
adamD = keras.optimizers.Adam(
    lr=1e-5, beta_1=0.5, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
adamG = keras.optimizers.Adam(
    lr=5e-6, beta_1=0.5, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
```

### Data Loading & Preprocessing

```python
bkg_all = pd.read_csv(bkg_filename, delimiter=' ', index_col=False)
sig_all = pd.read_csv(sig_filename, delimiter=' ', index_col=False)

def coordinate_change (df):
    pt1 = np.sqrt(df['px1']**2 + df['py1']**2)
    theta1 = np.arctan2(pt1,df['pz1'])
    eta1 = -1 * np.log(np.tan(theta1/2))
    phi1 = np.arctan2(df['py1'],df['px1'])
    pt2 = np.sqrt(df['px2']**2 + df['py2']**2)
    theta2 = np.arctan2(pt2,df['pz2'])
    eta2 = -1 * np.log(np.tan(theta2/2))
    phi2 = np.arctan2(df['py2'],df['px2'])
    df['px1'] = pt1
    df['py1'] = eta1
    df['pz1'] = phi1
    df['px2'] = pt2
    df['py2'] = eta2
    df['pz2'] = phi2
    df.rename(columns={'px1':'$p_{t1}$', 'py1':'$\eta_{1}$', 'pz1':'$\phi_{1}$', 'px2':'$p_{t2}$',
                       'py2':'$\eta_{2}$', 'pz2':'$\phi_{2}$', 'E1':'$E_1$', 'E2':'$E_2$', 'M1':'$M_1$',
                       'M2':'$M_2$', 'M12':'$M_{1 2}$'}, inplace=True)
    
coordinate_change(sig_all)
coordinate_change(bkg_all)

if drop_PHI_columns:
    sig_sel = sig_all.drop(columns = ['$\phi_{1}$', '$\phi_{2}$'])
    bkg_sel = bkg_all.drop(columns = ['$\phi_{1}$', '$\phi_{2}$'])
else:
    sig_sel = sig_all
    bkg_sel = bkg_all
    
data_header = list(sig_sel)

sig_shuffled = shuffle(sig_sel)
bkg_shuffled = shuffle(bkg_sel)

n_bkg = len(bkg_sel.iloc[:,:0])
n_sig = len(sig_sel.iloc[:,:0])
f_s = n_train/n_sig
f_b = n_train/n_bkg

print ("Number of BACKGROUND events:", n_bkg)
print ("%0.2f%% = %d" % (f_b*100, n_train), "used for TRAINING")
print ("%.2f%% = %d" % (f_b*100, n_train), "used for TESTING")
print ("%05.2f%% = %d" % ((1-2*f_b)*100, n_bkg - 2*n_train), "unused")
print ("Number of SIGNAL events:", n_sig)
print ("%.2f%% = %d" % (f_s*100, n_train), "used for TESTING")
print ("%.2f%% = %d" % ((1-f_s)*100, n_sig - n_train), "unused")
print ("\n")

sample_train = bkg_sel[:n_train]
sample_test = pd.concat([bkg_sel.iloc[n_train:2*n_train], sig_sel.iloc[:n_train]])
sample_test = sample_test.reset_index(drop=True)

input_dim = sample_train.shape[1]

sample_train = scaler.fit_transform(sample_train)
sample_test = scaler.transform(sample_test)

print('Training sample size: ',sample_train.shape)
print('Testing sample size: ',sample_test.shape)

sample_train_input, sample_train_valid = train_test_split(sample_train,test_size=0.2,random_state=13)
n_cycle=0
train_time=0
```

## Step 1: Train DCGAN
****

### Generator: CNN

```python
generator_in = Input(shape=(encoding_dim,), name='gen_in')
generator_f1 = Dense(128, activation=hidden_activation, name='gen_f1')(generator_in)
activate_f1 = LeakyReLU(alpha=leak)(generator_f1)
activate_f1 = BatchNormalization()(activate_f1)
generator_reshape = Reshape((8,8,2), name='generator_reshape')(activate_f1)
generator_c1 = Conv2DTranspose(64, kernel_size=2, strides=1, padding="same", name='gen_c1')(generator_reshape)
activate_gc1 = LeakyReLU(alpha=leak)(generator_c1)
activate_gc1 = BatchNormalization()(activate_gc1)
generator_c2 = Conv2DTranspose(32, kernel_size=3, strides=1, padding="same", name='gen_c2')(activate_gc1)
activate_gc2 = LeakyReLU(alpha=leak)(generator_c2)
activate_gc2 = BatchNormalization()(activate_gc2)
generator_c3 = Conv2DTranspose(16, kernel_size=3, strides=1, padding="same", name='gen_c3')(activate_gc2)
activate_gc3 = LeakyReLU(alpha=leak)(generator_c3)
activate_gc3 = BatchNormalization()(activate_gc3)
generator_f2 = Flatten()(activate_gc3)
generator_f3 = Dense(128, activation=hidden_activation, name='gen_f3') (generator_f2)
activate_gf4 = LeakyReLU(alpha=leak, name='extract')(generator_f3)
generated = Dense(input_dim, activation=generator_activation, name='gen_out')(activate_gf4)

generator = Model(generator_in, generated)
generator.summary()

class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
```

### Discriminator: CNN

```python
discriminator_in = Input(shape=(input_dim,), name='dis_in')
discriminator_expand = Dense(128, activation=hidden_activation, name='dis_expand')(discriminator_in)
activate_expand= LeakyReLU(alpha=leak)(discriminator_expand)
discriminator_reshape = Reshape((8,8,2), name='discriminator_reshape')(activate_expand)
discriminator_c1 = Conv2D(64, kernel_size=3, strides=1, padding="same", name='dis_c1')(discriminator_reshape)
activate_dc1 = LeakyReLU(alpha=leak)(discriminator_c1)
discriminator_c2 = Conv2D(32, kernel_size=3, strides=1, padding="same", name='dis_c2')(activate_dc1)
activate_dc2 = LeakyReLU(alpha=leak)(discriminator_c2)
discriminator_c3 = Conv2D(16, kernel_size=3, strides=1, padding="same", name='dis_c3')(activate_dc2)
activate_dc3 = LeakyReLU(alpha=leak)(discriminator_c3)
discriminator_f1 = Flatten()(activate_dc3)
discriminator_f2 = Dense(128, activation=hidden_activation, name='dis_f2') (discriminator_f1)
activate_df2 = LeakyReLU(alpha=leak)(discriminator_f2)
dis_out = Dense(2, activation=discriminator_activation, name='dis_out')(activate_df2)

discriminator = Model(discriminator_in, dis_out)
discriminator.compile(optimizer=adamD, loss='binary_crossentropy')
discriminator.summary()
```

### Adversarial Model

```python
AM_in = Input(shape=(encoding_dim,), name='AM_in')

generator_layers = generator.layers[17](generator.layers[16](generator.layers[15](generator.layers[14](generator.layers[13](generator.layers[12](generator.layers[11](generator.layers[10](generator.layers[9](generator.layers[8](
            generator.layers[7](generator.layers[6](generator.layers[5](generator.layers[4](
            generator.layers[3](generator.layers[2](generator.layers[1](AM_in)))))))))))))))))


discriminator_layers = discriminator.layers[13](discriminator.layers[12](discriminator.layers[11](discriminator.layers[10](discriminator.layers[9](discriminator.layers[8](
            discriminator.layers[7](discriminator.layers[6](discriminator.layers[5](discriminator.layers[4](
            discriminator.layers[3](discriminator.layers[2](discriminator.layers[1](generator_layers)))))))))))))

AM = Model (AM_in, discriminator_layers)

AM.get_layer('dis_expand').trainable = False
AM.get_layer('dis_c1').trainable = False
AM.get_layer('dis_c2').trainable = False
AM.get_layer('dis_c3').trainable = False
AM.get_layer('dis_f2').trainable = False
AM.get_layer('dis_out').trainable = False


AM.summary()
AM.compile(loss='binary_crossentropy', optimizer=adamG)
```

### Training

```python
train_steps
discriminator_history = []
adversarial_history = []

start = time.time()

Ghistory = LossHistory()
Dhistory = LossHistory()

with tf.device('/GPU:0'):
    for i in range(train_steps):
            data_train = sample_train[np.random.randint(0,sample_train.shape[0], size=batch_size_d)]
            noise = np.random.uniform(0, 1.0, size=[batch_size_d, encoding_dim])
            data_fake = generator.predict(noise)
            x = np.concatenate((data_train, data_fake))
            y = np.concatenate((np.tile([0,1], [batch_size_d,1]), (np.tile([1,0], [batch_size_d,1]))))
            d_train = discriminator.fit(x, y,
                    epochs=1,
                    batch_size=mini_batch,
                    shuffle=True,
                    callbacks=[Dhistory])
            discriminator_history.append(d_train.history['loss'][0])
            
            noise = np.random.uniform(0, 1.0, size=[batch_size_g, encoding_dim])
            y = np.tile([0,1], [batch_size_g,1])
            g_train =AM.fit(noise, y,
                    epochs=1,
                    batch_size=mini_batch,
                    shuffle=True,
                    callbacks=[Ghistory])
            adversarial_history.append(g_train.history['loss'][0])
            prc = ((i+1)/train_steps)*100
            print('\n')
            print(" - - - - - - - - - = = = = = | %.2f %% COMPLETED | = = = = = - - - - - - - - -" % prc)
            print('\n')
            
end = time.time()
train_time += (end-start)
n_cycle += 1
```

## OPTION 1: AEGAN

### Autoencoder

```python
encoder_in = Input(shape=(input_dim,), name='enc_in')
encoder_expand = Dense(128, activation=hidden_activation, name='enc_expand')(encoder_in)
activate_expand= LeakyReLU(alpha=leak)(encoder_expand)
encoder_reshape = Reshape((8,8,2), name='encoder_reshape')(activate_expand)
encoder_c1 = Conv2D(64, kernel_size=3, strides=1, padding="same", name='enc_c1')(encoder_reshape)
activate_ec1 = LeakyReLU(alpha=leak)(encoder_c1)
encoder_c2 = Conv2D(32, kernel_size=3, strides=1, padding="same", name='enc_c2')(activate_ec1)
activate_ec2 = LeakyReLU(alpha=leak)(encoder_c2)
encoder_c3 = Conv2D(16, kernel_size=3, strides=1, padding="same", name='enc_c3')(activate_ec2)
activate_ec3 = LeakyReLU(alpha=leak)(encoder_c3)
encoder_f1 = Flatten()(activate_ec3)
encoder_f2 = Dense(128, activation=hidden_activation, name='enc_f2') (encoder_f1)
activate_ef2 = LeakyReLU(alpha=leak)(encoder_f2)
encoded = Dense(encoding_dim, activation=encoder_activation, name='enc_out')(activate_ef2)

encoder=Model(encoder_in,encoded)

decoder_layers = generator.layers[17](generator.layers[16](generator.layers[15](generator.layers[14](generator.layers[13](generator.layers[12](generator.layers[11](generator.layers[10](generator.layers[9](generator.layers[8](
            generator.layers[7](generator.layers[6](generator.layers[5](generator.layers[4](
            generator.layers[3](generator.layers[2](generator.layers[1](encoded)))))))))))))))))

AE.get_layer('gen_f1').trainable = False
AE.get_layer('gen_c1').trainable = False
AE.get_layer('gen_c2').trainable = False
AE.get_layer('gen_c3').trainable = False
AE.get_layer('gen_f3').trainable = False
AE.get_layer('gen_out').trainable = False

AE = Model(encoder_in, decoder_layers)
AE.compile(optimizer=adamG, loss='mse')
AE.summary()
```

### Training

```python
n_epoch = train_steps//10

class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))

history = LossHistory()

start = time.time()
with tf.device('/GPU:0'):
    autoencoder_train = AE.fit(sample_train_input, sample_train_input,
                epochs=n_epoch,
                batch_size=mini_batch,
                shuffle=True,
                validation_data=(sample_train_valid, sample_train_valid),
                callbacks=[history])
end = time.time()
train_time2 = (end-start)
```

## OPTION 2: AnoGAN

### Load Discriminator

```python
if skipStep1:
    json_file = open('discriminator.json', 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    discriminator = model_from_json(loaded_model_json)
    discriminator.load_weights("discriminator.h5")
```

### Inverse Mapping

```python
L=0.1

encoder_in = Input(shape=(input_dim,), name='enc_in')
encoder_expand = Dense(128, activation=hidden_activation, name='enc_expand')(encoder_in)
activate_expand= LeakyReLU(alpha=leak)(encoder_expand)
encoder_reshape = Reshape((8,8,2), name='encoder_reshape')(activate_expand)
encoder_c1 = Conv2D(64, kernel_size=3, strides=1, padding="same", name='enc_c1')(encoder_reshape)
activate_ec1 = LeakyReLU(alpha=leak)(encoder_c1)
encoder_c2 = Conv2D(32, kernel_size=3, strides=1, padding="same", name='enc_c2')(activate_ec1)
activate_ec2 = LeakyReLU(alpha=leak)(encoder_c2)
encoder_c3 = Conv2D(16, kernel_size=3, strides=1, padding="same", name='enc_c3')(activate_ec2)
activate_ec3 = LeakyReLU(alpha=leak)(encoder_c3)
encoder_f1 = Flatten()(activate_ec3)
encoder_f2 = Dense(128, activation=hidden_activation, name='enc_f2') (encoder_f1)
activate_ef2 = LeakyReLU(alpha=leak)(encoder_f2)
encoded = Dense(encoding_dim, activation=encoder_activation, name='enc_out')(activate_ef2)

generator_layers = generator.layers[17](generator.layers[16](generator.layers[15](generator.layers[14](generator.layers[13](generator.layers[12](generator.layers[11](generator.layers[10](generator.layers[9](generator.layers[8](
            generator.layers[7](generator.layers[6](generator.layers[5](generator.layers[4](
            generator.layers[3](generator.layers[2](generator.layers[1](encoded)))))))))))))))))

discriminator_layers = discriminator.layers[12](discriminator.layers[11](discriminator.layers[10](discriminator.layers[9](discriminator.layers[8](
            discriminator.layers[7](discriminator.layers[6](discriminator.layers[5](discriminator.layers[4](
            discriminator.layers[3](discriminator.layers[2](discriminator.layers[1](discriminator.layers[0](generator_layers)))))))))))))

discriminator_only_layers = discriminator.layers[12](discriminator.layers[11](discriminator.layers[10](discriminator.layers[9](discriminator.layers[8](
            discriminator.layers[7](discriminator.layers[6](discriminator.layers[5](discriminator.layers[4](
            discriminator.layers[3](discriminator.layers[2](discriminator.layers[1](encoder_in))))))))))))

AS_out = [generator_layers, discriminator_layers]

DisExtract = Model(encoder_in, discriminator_only_layers)
InverseReco = Model(encoder_in, generator_layers)
AnomalyScore = Model(encoder_in, AS_out)

AnomalyScore.get_layer('gen_f1').trainable = False
AnomalyScore.get_layer('gen_c1').trainable = False
AnomalyScore.get_layer('gen_c2').trainable = False
AnomalyScore.get_layer('gen_c3').trainable = False
AnomalyScore.get_layer('gen_f3').trainable = False
AnomalyScore.get_layer('gen_out').trainable = False

AnomalyScore.get_layer('dis_expand').trainable = False
AnomalyScore.get_layer('dis_c1').trainable = False
AnomalyScore.get_layer('dis_c2').trainable = False
AnomalyScore.get_layer('dis_c3').trainable = False
AnomalyScore.get_layer('dis_f2').trainable = False

AnomalyScore.summary()

losses = {
    "gen_out": "mse",
    "leaky_re_lu_9": "mse",
}
lossWeights = {"gen_out": 1-L, "leaky_re_lu_9": L}

AnomalyScore.compile(optimizer=adamG, loss=losses, loss_weights=lossWeights)
AnomalyScore.summary()
```

### Training

```python
n_epoch = train_steps//10

class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []

    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))

history = LossHistory()

de_t = DisExtract.predict(sample_train_input)
de_v = DisExtract.predict(sample_train_valid)

start = time.time()
with tf.device('/GPU:0'):
    AS_train = AnomalyScore.fit(sample_train_input, [sample_train_input, de_t],
                epochs=n_epoch,
                batch_size=mini_batch,
                shuffle=True,
                validation_data=(sample_train_valid, [sample_train_valid, de_v]),
                callbacks=[history])

end = time.time()
train_time2 = (end-start)
```

### Loss

```python
loss = AS_train.history['loss']
val_loss = AS_train.history['val_loss']
Rloss = AS_train.history['gen_out_loss']
val_Rloss = AS_train.history['val_gen_out_loss']
Dloss = AS_train.history['leaky_re_lu_9_loss']
val_Dloss = AS_train.history['val_leaky_re_lu_9_loss']
nepochs = range(n_epoch)
lossFig = plt.figure(figsize=(18,6))
plt.rcParams.update({'font.size': 16})
plt.plot(nepochs, loss, 'g', label='Training loss')
plt.plot(nepochs, val_loss, 'go', label='Validation loss')
plt.plot(nepochs, Rloss, 'b', label='Training loss')
plt.plot(nepochs, val_Rloss, 'bo', label='Validation loss')
plt.plot(nepochs, Dloss, 'r', label='Training loss')
plt.plot(nepochs, val_Dloss, 'ro', label='Validation loss')
plt.title('Training and validation loss')
plt.yscale('log')
plt.legend()
plt.show()
```

### Anomaly Score Calculation

```python
test_as = AnomalyScore.predict(sample_test)
train_as = AnomalyScore.predict(sample_train)
test_d = DisExtract.predict(sample_test)
train_d = DisExtract.predict(sample_train)

norm_bkg_test=[]
norm_bkg_test_r=[]
norm_bkg_test_d=[]
for i in range(sample_test.shape[0]//2):
    r = norm(sample_test[i,:]-test_as[0][i,:])
    d = norm(test_d[i,:]-test_as[1][i,:])
    norm_bkg_test.append(L*d+(1-L)*r)
    norm_bkg_test_r.append(r)
    norm_bkg_test_d.append(d)

norm_sig=[]
norm_sig_r=[]
norm_sig_d=[]
for i in range(sample_test.shape[0]//2, sample_test.shape[0]):
    r = norm(sample_test[i,:]-test_as[0][i,:])
    d = norm(test_d[i,:]-test_as[1][i,:])
    norm_sig.append(L*d+(1-L)*r)
    norm_sig_r.append(r)
    norm_sig_d.append(d)
    
norm_bkg_train=[]
norm_bkg_train_r=[]
norm_bkg_train_d=[]
for i in range(sample_train.shape[0]):
    r = norm(sample_train[i,:]-train_as[0][i,:])
    d = norm(train_d[i,:]-train_as[1][i,:])
    norm_bkg_train.append(L*d+(1-L)*r)
    norm_bkg_train_r.append(r)
    norm_bkg_train_d.append(d)
    

    
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
n, bins, patches = plt.hist(norm_sig, bins='auto', alpha=0.5, label='SIG', density=True)
plt.hist(norm_bkg_train, bins, alpha=0.5, label='BKG train', density=True)
plt.hist(norm_bkg_test, bins, alpha=0.5, label='BKG test', density=True)
plt.xlabel('Anomaly Score')
plt.legend(loc='upper right')

plt.subplot(1, 3, 2)
n, bins, patches = plt.hist(norm_sig_r, bins='auto', alpha=0.5, label='SIG', density=True)
plt.hist(norm_bkg_train_r, bins, alpha=0.5, label='BKG train', density=True)
plt.hist(norm_bkg_test_r, bins, alpha=0.5, label='BKG test', density=True)
plt.xlabel('Reco Loss')
plt.legend(loc='upper right')

plt.subplot(1, 3, 3)
n, bins, patches = plt.hist(norm_sig_d, bins='auto', alpha=0.5, label='SIG', density=True)
plt.hist(norm_bkg_train_d, bins, alpha=0.5, label='BKG train', density=True)
plt.hist(norm_bkg_test_d, bins, alpha=0.5, label='BKG test', density=True)
plt.xlabel('Discr Loss')
plt.legend(loc='upper right')


if savePlots:
    plt.savefig(plotsLocation + "distances.png")
```

### ROC

```python
target = np.concatenate((np.zeros(sample_train.shape[0] + sample_test.shape[0]//2),np.ones(n_train)))
scores = np.concatenate((norm_bkg_train,norm_bkg_test, norm_sig))  

train_time_t=train_time+train_time2

train_min = int(train_time_t) // 60
train_sec = train_time_t - 60 * train_min
time_stamp = "Total GPU train time: %d'  %.1f''" % (train_min, train_sec)

plt.figure(figsize=(7, 7))
                   
fp, vp, thresholds = roc_curve(target,scores)
roc_auc = auc(fp, vp)

plt.plot(fp,vp,color='red',label='ROC curve %s (AUC = %0.4f)'%('AE',roc_auc))
plt.xlabel('FP')
plt.ylabel('TP')
plt.plot([0, 1],[0, 1], linestyle='--',color=(0.6, 0.6, 0.6), label='Random guess')
plt.grid()
plt.legend(loc="best", title =time_stamp)
plt.tight_layout()

if savePlots:
    plt.savefig(plotLocation + "ROC.png")
```
